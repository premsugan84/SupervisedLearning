# KMeans Clustering

```python
#Now lets plot the elbow matrix to find out the optimal number of clusters 
from scipy.spatial.distance import cdist
clusters=range(1,10)
meandistortions=[]

for k in clusters :
    model = KMeans(n_clusters=k)
    model.fit(cc_df_scaled)
    prediction = model.predict(cc_df_scaled)
    meandistortions.append(sum(np.min(cdist(cc_df_scaled, model.cluster_centers_, 'euclidean'), axis=1))/cc_df_scaled.shape[0])
    
plt.plot(clusters, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average Distortion')
plt.title('Selecting k with elbow method')
```

```python
# It has a definite elbow at 3 and slight elbow at 2. For further plotting we will take k= 3
final_model = KMeans(3)
final_model.fit(cc_df_scaled)
prediction = final_model.predict(cc_df_scaled)

#Append the prediction
cc_df_scaled["Group"] = prediction
cc_df["Group"] = prediction
cc_df.head()
#So it is evident we got three groups from the data
```

```python
# lets calcualte the mean to find out the average that falls in each group 
cc_df_clust = cc_df.groupby(["Group"])
cc_df_clust.mean()
#From the below mean it states that the group 2 which has customers wtih more credit cards has more credit limit and does
#like to do most transactions online
#In Group 1 which has less credit limit than 2 but more than 1 has average of 5 credit cards is a different set of customers
# who like to visit bank more than doing transactions online
#Group 2 however does like to call and do their transactions or interactions more with the card company than visiting and online
```

# Hiearchial Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')
```

```python
#We will split the data into three clusters
model.fit(cc_df_scaled)
```

```python
cc_df['Group'] = model.labels_
cc_df.head()
#We are able to see the data got seperated into three groups
```

```python
cc_df_clust = cc_df.groupby(['Group'])
cc_df_clust.mean()
```

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage 
```

```python
from scipy.spatial.distance import pdist
```

```python
z = linkage(cc_df_scaled, metric='euclidean', method='average')
c, coph_dists = cophenet(z, pdist(cc_df_scaled))

c
#Being 1 is the ideal coefficent, 0.89 seems to be a pretty good determing the quality of the clusters
```

```python
plt.figure(figsize=(10,5))
plt.title('Agglomerative Hierarchial Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(z, leaf_rotation=90., color_threshold=40, leaf_font_size=8.)
plt.tight_layout()
# When we draw a little over 3 we get three clusters. Based om the clusters we need we draw an horizontal line which will 
# give us the required number of clusters
```

# Silhouette analysis

```python
# We are going to try with the complete linkage and see if there is a coefficent difference
z = linkage(cc_df_scaled, metric='euclidean', method='complete')
c, coph_dists = cophenet(z, pdist(cc_df_scaled))

c
#Being 1 is the ideal coefficent, 0.85  though seems to be a good it is lesser than the average linkage 
```

```python
plt.figure(figsize=(10,5))
plt.title('Agglomerative Hierarchial Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(z, leaf_rotation=90., color_threshold=40, leaf_font_size=8.)
plt.tight_layout()
# The dendogram looks like little different from distance and this linkage will be helpful if there is a need between
# 6 to 10 clusters
```

```python
# We are going to try with the complete linkage and see if there is a coefficent difference
z = linkage(cc_df_scaled, metric='euclidean', method='ward')
c, coph_dists = cophenet(z, pdist(cc_df_scaled))

c
#Being 1 is the ideal coefficent, 0.74 is lesser than both average and complete
```

```python
plt.figure(figsize=(10,5))
plt.title('Agglomerative Hierarchial Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(z, leaf_rotation=90., color_threshold=40, leaf_font_size=8.)
plt.tight_layout()
```

Out of all three linkages that are tried, average linkage seems to be better option of all three so for this data set Average linkage is the best suitable cluster group

```python
vehicle_df.shape
```

```python
#Seperating the dependent and independent variable. Dependent variable is class 
x = vehicle_df.drop(['class'], axis=1)
y = vehicle_df[['class']]
```

```python
sns.pairplot(x, diag_kind='kde')
#From the pairplot circularity, scatter ratio, elongatness, radius of gyration have strong realtionship
```

```python
# From the pairplot aspect ratio and rectangularity between the axis and max length shows the same pattern so will drop
#the max length and keep the axis 
x = x.drop(['max.length_rectangularity'], axis=1)
x = x.drop(['max.length_aspect_ratio'], axis=1)
```

```python
x.shape
```

```python
#Lets apply the score to remove the units
from scipy.stats import zscore 
XScaled = x.apply(zscore)
XScaled.head()
```

```python
covMatrix = np.cov(XScaled, rowvar=False)
print(covMatrix)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=11)
pca.fit(XScaled)
```

The eigen values 

```python
print(pca.explained_variance_)
```

The eigen vectors

```python
print(pca.components_)
```

And the percentage of the variation explained by eigen vector

```python
print(pca.explained_variance_ratio_)
# As we can see after the first four the change is minimal
```

```python
#Plot to see the variance with the eigen value 
plt.bar(list(range(1,12)),pca.explained_variance_ratio_,alpha=0.5, align='center')
plt.ylabel('Variation explained')
plt.xlabel('eigen value')
plt.show()
```

```python
#Plotting to find the optimal no of components for 95% variance 
plt.step(list(range(1,12)),np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Cum of Variation explained')
plt.xlabel('eigen value')
plt.show()
```

# Dimensionality reduction

```python
#Since 4 is covering more than 95% of the variance. we will do the PCA again with 4 components 
pca_new = PCA(n_components=4)
pca_new.fit(XScaled)
```

The eigen vectors

```python
print(pca_new.components_)
```

The eigen values

```python
print(pca_new.explained_variance_)
```

```python
Xpca_new = pca_new.transform(XScaled)
sns.pairplot(pd.DataFrame(Xpca_new))
```
